```python
##################################################################################################################################################################
# set result directory
##################################################################################################################################################################

RES_PATH = {
    'PERF_ROOT': f'result/performance/',
    'PRED_ROOT': f'result/prediction/',
    'ML':        f'ml/',
    'CDDA':      f'cdda/',
    # 'CDDA_DHY':  f'cdda/DHY/',
    # 'CDDA_GJK':  f'cdda/GJK/'
}

res_dir_pred = RES_PATH['PRED_ROOT'] + RES_PATH['CDDA']
res_dir_perf = RES_PATH['PERF_ROOT'] + RES_PATH['CDDA']

file_name    = '250615_APP_PDX_SGD_MRDDM_REC'
```

```python
res_df_pred = pd.read_csv(f'{res_dir_pred}{file_name}_V1_PRED.csv')
res_df_perf = pd.read_csv(f'{res_dir_perf}{file_name}_V1_PERF.csv')

target      = 'Target'
init_tr_end_idx  = res_df_perf['init_tr_end_idx'][0]
mdl_upd_idx_list = ast.literal_eval(res_df_perf['cd_idx'][0])
res_df_pred['time'] = pd.to_datetime(res_df_pred['time_idx'])

plt.figure(figsize=(10, 4))
plt.plot(res_df_pred['time'], res_df_pred['y_real'], label='Real', color='black')
plt.plot(res_df_pred['time'], res_df_pred['y_pred'], label='Pred', color='red')
plt.legend()
plt.ylabel(target)
for idx in mdl_upd_idx_list:
    plt.axvline(x=res_df_pred['time'][idx-init_tr_end_idx], linestyle='--')
# for
plt.show()
```
